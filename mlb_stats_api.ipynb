```python
# trying to scrape data from a game today, July 8 2022, Rays @ Reds (game id 663053)
# api endpoint is playByPlay

url = 'https://statsapi.mlb.com/api/v1/game/663053/playByPlay'
res = requests.get(url)
# turn the response into JSON
data = res.json()
type(data)
```

```python
for i in range(len(data['allPlays'])):
    print(data["allPlays"][i]["matchup"]["batter"]["fullName"], data['allPlays'][i]['result']['event'])
    event = data['allPlays'][i]['result']['event']
    if event == 'Single' or event == 'Double' or event == 'Triple' or event == 'Home Run':

        # define 'name' to be in proper format for baseball reference game log url.  ie 'troutmi'
        name = data["allPlays"][i]["matchup"]["batter"]["fullName"]
        name_bref = name.split()
        name_bref = name_bref[1][:5] + name_bref[0][:2]
        name_bref = name_bref.lower()
        print(f'we have a hit by {name}, or {name_bref}')
      

        # access player game log and check how long streak is 
        bref_url = f'https://www.baseball-reference.com/players/gl.fcgi?id={name_bref}01&t=b&year=2022'
        print(bref_url)
        df_gamelog = pd.read_html(bref_url)
        df_gamelog = pd.DataFrame(df_gamelog[4])
        df_gamelog = df_gamelog.loc[df_gamelog['H'] != 'H']
        print(df_gamelog.shape)
           
        # check how long player's current hit streak is
        hit_streak = 0
        for val in df_gamelog.sort_index(ascending=False)[1:]['H']:
            if int(val) > 0:
                hit_streak += 1                
            else:
                print(f'{name} has a {hit_streak} game hit streak going')
                break
                    
        break
```
